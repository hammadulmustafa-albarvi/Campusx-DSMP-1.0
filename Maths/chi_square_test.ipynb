```python
import numpy as np
import seaborn as sns
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
sample1 = np.random.normal(loc=0,scale=1,size=100)
sample2 = np.random.normal(loc=0,scale=1,size=100)
sample3 = np.random.normal(loc=0,scale=1,size=100)
sample4 = np.random.normal(loc=0,scale=1,size=100)
sample5 = np.random.normal(loc=0,scale=1,size=100)

# standard normal variate where mean = 0 and std = 1
```

```python
x = sample1**2
y = sample1**2 + sample2**2
z = sample1**2 + sample2**2 + sample3**2
c = sample1**2 + sample2**2 + sample3**2 + sample4**2
v = sample1**2 + sample2**2 + sample3**2 + sample4**2 + sample5**2


```

```python
sns.kdeplot(x,clip=(x.min(),x.max()))
sns.kdeplot(y,clip=(y.min(),y.max()))
sns.kdeplot(z,clip=(z.min(),z.max()))
sns.kdeplot(c,clip=(c.min(),c.max()))
sns.kdeplot(v,clip=(v.min(),v.max()))
```

# goodness of fit test

```python
df = sns.load_dataset('titanic')
```

```python
df['pclass'].value_counts()
```

```python
#  pclass
# 3	491
# 1	216
# 2	184
# here we have more people in pclass 3 than in rest but our expected data would be
# total * probablity of selecting one value (expected value)
# 891 * 1/3 = 297

# calculate chi square statistic by (original value - expected value)^2 / expected value

# (491-297)^2 /297 + (216-297)^2 /297 + (184-297)^2 /297

```

```python
from scipy.stats import chisquare
chi2,p_value = chisquare(df['pclass'].value_counts().sort_index(),df['pclass'].value_counts().sum()/3)
```
