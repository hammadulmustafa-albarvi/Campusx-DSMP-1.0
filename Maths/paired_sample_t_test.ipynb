```python
# compare variance of both data
from scipy.stats import levene
levene_test = levene(desktop_users,mobile_users)
levene_test
```

# Case study on titanic

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
pop_male = df[df['sex']=='male']['age'].dropna()
pop_female = df[df['sex']=='female']['age'].dropna()
```

```python
sample_male = pop_male.sample(25)
sample_female = pop_female.sample(25)
```

```python
alpha = 0.05
```

```python
# Ho average age of female = average age of male
# Ha average age of female < average age of male
# alpha = 0.05
```

```python
from scipy.stats import shapiro
shapiro(sample_male)
```

```python
shapiro(sample_female)
```

```python
# since both p_values of shapiro test > 0.05 than
# so it means they are normal distribution

```

```python
from scipy.stats import levene
levene(sample_female,sample_male)
```

```python
# again p_value > 0.05 so there variance is same
```

```python
 import scipy.stats as stats
 t_statistic,p_value = stats.ttest_ind(sample_female,sample_male)
 print(t_statistic)
 print(p_value/2)
```

```python
# since p_value > 0.05 than we cant reject null hypothesis
```
