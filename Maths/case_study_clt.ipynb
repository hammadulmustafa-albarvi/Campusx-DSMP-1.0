```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = sns.load_dataset('titanic')
```

```python
df
```

```python
df['fare'].plot(kind='kde')
```

```python
 df = df.sample(891)
```

```python
sample_size = 50
number_samples = 100
```

```python
sample_mean = []
```

```python
for i in range(number_samples):
  sample_df  = df.sample(sample_size)
  sample_mean.append(sample_df['fare'].mean())
```

```python
pd.DataFrame(sample_mean)
```

```python
pd.DataFrame(sample_mean).plot(kind='kde')
```

```python
pd.DataFrame(sample_mean).mean()
```

```python
df['fare'].mean()
```

```python
df['fare'].var()/100
```

```python
pd.DataFrame(sample_mean).std()/np.sqrt(50)
```

```python
lower_limit = pd.DataFrame(sample_mean).mean() - 2*pd.DataFrame(sample_mean).std()/np.sqrt(50)
upper = pd.DataFrame(sample_mean).mean() + 2*pd.DataFrame(sample_mean).std()/np.sqrt(50)
```

```python
print(f'{lower_limit} - {upper}')
```
