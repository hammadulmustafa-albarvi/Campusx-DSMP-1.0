Consider this "wine" dataset which can be loaded using above code. It has 13 numerical features and a target column with 3 class: 0, 1, 2.


* **Which feature is the best predictor of the wine class in the wine dataset?** (Using pdf plot)

* **Which two combination of features from below listed features are best for wine classification?** (2D pdf plot)
    * Proline
    * Flavanoids
    * Color intensity
    * OD280/OD315 of diluted wines
    * Alcohol

````python
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
# Load dataset
import pandas as pd
from sklearn.datasets import load_wine

# load the wine dataset
data = load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)
df["target"] = data.target
````

````python
for i in df.columns:
  plt.figure()
  sns.kdeplot(df,hue='target',x=i)
  plt.title(f'{i}')
  plt.show()
````

````python
selected_features = df.columns[[0,6,9,11,12]]
for i in selected_features:
  for j in selected_features:
    if i != j: # Avoid plotting the same feature against itself
      g = sns.jointplot(data=df, x=i, y=j, kind='kde', fill=True, cbar=True, hue='target')
      g.fig.suptitle(f'Joint KDE plot of {i} vs {j}', y=1.02) # Add title to joint plot
      plt.show()
````
