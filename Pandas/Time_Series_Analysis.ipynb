```python
# REAL EXAMPLE
import pandas as pd

deliveries = pd.DataFrame({
    'order_date': pd.to_datetime(['2025-07-01', '2025-07-05', '2025-07-10']),
    'delivery_date': pd.to_datetime(['2025-07-04', '2025-07-08', '2025-07-13'])
})


```

```python
deliveries.head()
```

```python
deliveries['order_date'] = pd.to_datetime(deliveries['order_date'])
```

```python
deliveries['delivery_date'] = pd.to_datetime(deliveries['delivery_date'])
```

```python
deliveries.info()
```

```python
deliveries['delivery_time_period']=deliveries['delivery_date'] - deliveries['order_date']
```

```python
deliveries.head()
```

```python
# weekly trend
google.groupby('day_name')['Close'].mean().plot(kind='bar')
```

```python
# monthly trend
google.groupby('month_name')['Close'].mean().plot(kind='bar')
```

Resampling

```python
#down sampling
google['Close'].plot()
google['Close'].resample('Y').mean().plot()
```

```python
 #upsampling
 google['Close'].resample('H').interpolate(method='linear').plot()
```

Rolling Window ( Smoothing )

```python
# calculate moving average than make plot
# google['Close'].plot()
google['Close'].rolling(20).mean().plot()
google['Close'].ewm(20).mean().plot()
```
