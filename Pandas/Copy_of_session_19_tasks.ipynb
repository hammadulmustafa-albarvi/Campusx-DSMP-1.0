### `Q-2:` Using `groupby` make groups using the `"Pclass"` column and fill every group's `"Embarked"` column's missing values with the mode value of that group. After that, print every group's `"Embarked"` column's value counts in ascending order.

````python
# code here
for i in sorted(df['Pclass'].unique()):
  df.groupby('Pclass').get_group(i)['Embarked'].fillna(df.groupby('Pclass').get_group(i)['Embarked'].mode())
  print(f"Pclass : {i}\n{df[df['Pclass']==i]['Embarked'].value_counts()}")
````

### `Q-4:` Perform `groupby` based on the `"Team"` column and then perform Z Normalization on top of the below columns of each group:
1. Passes
2. Passes Completed
3. Attempted Line Breaks
4. Completed Line Breaks

You have to make a python function named `z_normalization` which takes two arguments:

1. *group:* Every group that you have created
2. *cols_to_perform:* This parameter takes a list of columns on which you have to perform the Z-Normalization.

For this problem, you have to use th `apply()` method.

$\Large Z \ - \ Normalization = \frac{X_i - \mu}{std} $

After that find out the below values for each group:
- minimum "Passess"
- maximum "Passess"
- minimum "Yellow Cards"
- maximum "Yellow cards"
- average "Yellow Cards"
- maximum "Attempted Line Breaks"
- minimum "Attempted Line Breaks"
- standardard deviation of "Attempted Line Breaks"
- average Possession

````python
fifa.groupby('Team').agg({
    'Passes' : ['min','max'],
    'Yellow Cards' : ['min','max','mean'],
    'Attempted Line Breaks' : ['max','min','std'],
    'Possession (%)' : 'min'
})
````

````python
cols_list = list(('Passes','Passes Completed','Attempted Line Breaks','Completed Line Breaks'))
def z_normalization(group,cols_list):
  for i in cols_list:
    group[i] = (group[i] - group[i].mean()) / group[i].std()
  return group
fifa.groupby('Team').apply(z_normalization,cols_list).head(1)
````

###`Q-5:` Find batsman in below category-
* Highest score while chasing
* Best Strike rate while chasing and have faced 100+ balls


> Chasing mean team batting in second inning

````python
ipl = ipl[~(ipl['extra_type'] == 'wides')]
data = ipl.groupby('batter').agg({
    'batsman_run' : 'sum',
    'ballnumber' : 'count'
})
data = data[data['ballnumber'] >= 100]
result = data['batsman_run']/data['ballnumber'] * 100
result.sort_values(ascending=False).head()
````

###`Q-7`: Most successful batting pair in IPL. Batting pair who have scored most runs playing together.


````python
def func(g):
  return '-'.join(list(np.sort(g.values)))

ipl['partner_ship'] = ipl[['batter','non-striker']].apply(func,axis=1)
ipl.groupby('partner_ship')['total_run'].sum().sort_values(ascending=False)
````

###`Q-8:` Make a dataframe for all batting pairs played together.
```
Batsman1 Batsman2 Runs Avg StrikeRate
```

> Just to ease this question you can count wide-balls for strike rate.

````python
ans = []
colums = ['Batsman1','Batsman2','Runs','Avg','Strikerate']
for i in ipl['partner_ship'].unique():
  data = []
  total_runs = ipl[ipl['partner_ship']==i].groupby('partner_ship')['total_run'].sum().values[0]
  average_run = total_runs / (ipl[ipl['partner_ship']==i].groupby('partner_ship')['player_out'].count().values[0])
  strike_rate =( total_runs / (ipl[ipl['partner_ship']=='AB de Villiers-V Kohli'].groupby('partner_ship')['ballnumber'].count().values[0])) * 100
  try:
    batsman1,batsman2 = i.split('-')
    data.extend([batsman1,batsman2,total_runs,average_run,strike_rate])
    ans.append(data)
  except ValueError:
    continue


result = pd.DataFrame(data=ans,columns=colums)
result.head()
````

````python
result = result.sort_values(by=['Strikerate','Avg','Runs'],ascending=[False,False,False])
````
