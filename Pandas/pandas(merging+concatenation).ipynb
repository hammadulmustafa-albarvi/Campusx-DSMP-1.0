# ***Merge***

```python
students.head()
```

```python
regs.head()
```

```python
# students is left table regs is right table we will do inner join which will find common elements between 2 on basis of student_id
students.merge(regs,how='inner',on='student_id')
```

```python
# here left table is courses and right table is regs we do lef join in which we will get all contents of left table even those which arent associated with any student value
courses.merge(regs,how='left',on='course_id')
```

```python
# here left table is courses and right table is regs we do right join in which we will get all contents of right table even those which arent in left table
students.merge(regs,how='right',on='student_id')
```

```python
# outer join will show all data of both table , common elements too  whether it has nan values or not
students.merge(regs,how='outer',on='student_id')
```

# **FIND SIXES PER VENUE RATIO**

```python
matches.head()
```

```python
delivieries.head()
```

```python
delivieries.columns
```

```python
ipl = delivieries.merge(matches,left_on='match_id',right_on='id')
```

```python
sixes = ipl[ipl['batsman_runs']==6].groupby('venue')['venue'].count()
```

```python
matchs = matches['venue'].value_counts()
```

```python
# sixes per venue
(sixes/matchs).sort_values(ascending=False).head(3)
```

```python
ipl.groupby(['season','batsman'])['batsman_runs'].sum().reset_index().sort_values('batsman_runs',ascending=False).drop_duplicates(subset=['season'],keep='first')
```
