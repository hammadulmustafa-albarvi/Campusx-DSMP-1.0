```python
movies.groupby('Genre').agg({
    'Runtime' : 'mean',
    'IMDB_Rating' : 'mean',
    'No_of_Votes' : 'sum',
    'Gross' : 'sum',
    'Metascore':'min'}

)
```

```python
movies.groupby('Genre')[movies.select_dtypes(include=np.number).columns].agg(['min','max','mean'])
```

```python
movies.groupby('Genre').agg({
    'Runtime' : ['min','mean'],
    'IMDB_Rating' : ['min','mean'],
    'No_of_Votes' : 'sum',
    'Gross' : 'sum',
    'Metascore':['min','mean']}

)
```

```python
# highest rated movie of each genre
for group,data in movies.groupby('Genre'):
  print(data[data['IMDB_Rating'] == data['IMDB_Rating'].max()])
```

```python
# apply
movies.groupby('Genre').apply(min)
```

```python
#no of moviews in each genre which start with A
def check_a(group):
  return group['Series_Title'].str.startswith('A').sum()

movies.groupby('Genre').apply(check_a)
```

```python
# assigning rank to movie on baisis of imdb rating of each genre

def ranking(g):
  g['movie_ranking'] =  g['IMDB_Rating'].rank(ascending=False)
  return g

movies.groupby('Genre').apply(ranking)
```

```python
def normalize(g):
  g['Normalize_rating'] = (g['IMDB_Rating']-g['IMDB_Rating'].min())/(g['IMDB_Rating'].max()-g['IMDB_Rating'].min())
  return g

movies.groupby('Genre').apply(normalize)
```

```python
# group by on multiple columns
movies.groupby(['Director','Star1']).size()
movies.groupby(['Director','Star1']).get_group(("Aamir Khan","Amole Gupte"))
```

```python
# find most earning actor -> director combo
movies.groupby(['Director','Star1'])['Gross'].sum().sort_values(ascending=False).head(1)
```

```python
# find best actor -> genre combo in term of meta score
movies.groupby(['Star1','Genre'])['Metascore'].mean().reset_index().sort_values('Metascore',ascending=False).head(1)
```

```python
# perfrom aggregate on groupby of multiple columns
movies.groupby(['Star1','Genre']).agg({
    'IMDB_Rating' : 'mean',
    'Runtime' : 'mean'}
)
```

```python
movies.groupby(['Star1','Genre']).agg(['min','max'])
```

```python
ipl = pd.read_csv("/content/deliveries.csv")
```

```python
ipl.head(1)
```

```python
# top 10 batsman in terms of runs scored
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).reset_index().head(1)
```

```python
# find batsman with max number of sixes
ipl[ipl['batsman_runs'] == 6].groupby('batsman')['total_runs'].count().sort_values(ascending=False)
```

```python
# find batsman who did most 4's and 6's in last 5 over
ipl[(ipl['over'] > 15) & ((ipl['batsman_runs'] == 6) | (ipl['batsman_runs'] == 4))].groupby('batsman')['total_runs'].count().sort_values(ascending=False)
```

```python
# find v.kohlis record agains all teams
ipl[ipl['batsman'] == 'V Kohli'].groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
```

```python
# virat kohli highest score
ipl[ipl['batsman'] == 'V Kohli'].groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(1)
```

```python
# find highest score in single match
ipl[ipl['batsman'] == 'CH Gayle'].groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1)
```
