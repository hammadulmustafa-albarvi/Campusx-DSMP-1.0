```python
# how to create multi index object
# 1. pd.MultiIndex.from_tuples()

index_val = [('cse',2019),('cse',2020),('cse',2021),('ece',2019),('ece',2020)]
multiindex = pd.MultiIndex.from_tuples(index_val)
multiindex
```

```python
#  2. multiindex.from_product

p = pd.MultiIndex.from_product([['cse','ece'],[2019,2020,2021,2022]])
p
```

```python
#create a multi index series

m = pd.Series([1,2,3,4,5,6,7,8],index=p)
```

```python
 # how to fetch item from multi index series

m['cse']
m[('cse',2020)]
```

```python
m.ndim
```

```python
# unstack
us = m.unstack()
us
```

```python
# stack
s = us.stack()
s.ndim
s
```
