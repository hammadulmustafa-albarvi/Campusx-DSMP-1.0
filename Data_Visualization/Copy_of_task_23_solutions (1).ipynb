### **`Problem-1:`** Draw a line plot of which, the x-axis is the "Year" and the y-axis is sum of "PM2.5" of two countries Iran and China.

```python
# code here
df = pd.read_csv('https://tinyurl.com/2fe6vz4u')
df.head()
```

```python
plt.plot(df[df['Country']=='China'].groupby('Year')['PM2.5'].sum(),label='China')
plt.plot(df[df['Country']=='Iran'].groupby('Year')['PM2.5'].sum(),label='Iran')
plt.title("Iran and China PM2.5 by year comparision")
plt.xlabel("YEAR")
plt.ylabel("PM2.5")
plt.legend()
```

### **`Problem-2:`** Draw a histogram of the  column "PM10" of which the y-axis represents the probability (see the documentation how to draw the probability).

```python
plt.hist(df['PM10'],density=True)
plt.title('Probabilty distribution of PM10')
plt.xlabel("PM10 values")
plt.ylabel('probability')
```

### **`Problem-3:`** Draw a scatter plot where x-axis represents "PM2.5" and y-axis represents "PM10" for two countries Poland and Chile.

```python
plt.scatter(df[df['Country'] =='Poland']['PM2.5'],df[df['Country'] =='Poland']['PM10'],label='Poland')
plt.scatter(df[df['Country'] =='Chile']['PM2.5'],df[df['Country'] =='Chile']['PM10'],label='Chile')
plt.title('Scatter plot')
plt.xlabel('PM2.5')
plt.ylabel('PM10')
plt.legend()
```

###`Problem-9` Show Quarter wise Sales data for all products as multi Bar chart.


```python
df[df['month_number']<=3]
```

```python
df[(df['month_number']>=3) & (df['month_number']<=6)]
```

```python
df['quarter'] = pd.to_datetime([f'2020-{i}-01' for i in df['month_number']])
```

```python
data = df.groupby(df['quarter'].dt.quarter).sum(numeric_only=True)
```

```python
n = len(data.columns[1:7])
x = np.arange(data.shape[0])
bar_width = 0.15
for i,col in enumerate(df.columns[1:7]):
  offset = (i-(n-1)/2) * bar_width
  plt.bar(offset+x,data[col],width=bar_width)

```

###`Problem-10` Plot Stacked Bar chart quarter wise for each product.

```python
data
```

```python
prev=None
for i in data.columns[1:7]:
  if prev is not None:
    plt.bar(data['month_number'],data[i],bottom=prev,label=i,width=3.5)
    prev += data[i]
  else:
    plt.bar(data['month_number'],data[i],label=i,width=3.5)
    prev=data[i]


plt.legend()
```
