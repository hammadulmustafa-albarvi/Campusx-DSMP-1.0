```python
batter = pd.read_csv('/content/batter.csv')
```

```python
sample_df = batter.head(100).sample(25)
```

```python
sample_df
```

```python
plt.figure(figsize=(15,9))

plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
plt.axhline(130,color='red')
plt.axvline(30,color='red')
plt.axhline(135,color='green')
for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i],fontdict={'size':12,'color':'brown'})
```

SubPlots

```python
fig,ax = plt.subplots(figsize=(15,7))

ax.scatter(batter['avg'],batter['strike_rate'])
ax.set_title('Scatter Plot')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike rate')
fig.show()
```

```python
fig ,ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))
ax[0].scatter(batter['avg'],batter['strike_rate'],c='red')
ax[1].scatter(batter['avg'],batter['runs'])
ax[0].set_title('Avg vs Strike rate')
ax[1].set_title('Avg vs Runs')
```

```python
fig ,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,6))
ax[0,0].scatter(batter['avg'],batter['strike_rate'],c='red')
ax[0,1].scatter(batter['avg'],batter['runs'])

ax[1,0].hist(batter['runs'])
ax[1,1].hist(batter['strike_rate'])
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.scatter(batter['avg'],batter['strike_rate'],c='red')
ax2 = fig.add_subplot(1,2,2)
ax2.hist(batter['runs'])
```

3D Scatter Plot

```python
fig = plt.figure()
ax=plt.subplot(projection='3d')
ax.scatter3D(batter['runs'],batter['strike_rate'],batter['avg'],marker='+')
```

3D line plot

```python
x=np.arange(-10,10,0.5)
y=np.arange(-10,10,0.5)
```

```python
y
```

```python
xx,yy = np.meshgrid(x,y)
```

```python
z = xx**2+yy**2
```

```python
fig = plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')
P = ax.plot_surface(xx,yy,z,cmap='viridis')
plt.colorbar(P)
```

Contour Plot

```python
fig = plt.figure(figsize=(10,8))
ax=plt.subplot()
P = ax.contour(xx,yy,z,cmap='viridis')
plt.colorbar(P)
```

```python
fig = plt.figure(figsize=(10,8))
ax=plt.subplot()
P = ax.contourf(xx,yy,z,cmap='viridis')
plt.colorbar(P)
```
