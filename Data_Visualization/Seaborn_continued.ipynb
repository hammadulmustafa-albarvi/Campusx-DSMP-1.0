```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

## Categorical Plots

### Categorical Scatter Plot

- Stripplot
- Swarmplot

### Categorical Distribution Plots

- Boxplot
- Violinplot

### Categorical Estimate Plot -> for central tendency

- Barplot
- Pointplot
- Countplot

### Figure level function -> `catplot`

CATEGORICAL SCATTER PLOT

```python
sns.scatterplot(tips,x='total_bill',y='tip')
```

```python
 # strip plot -> axis level
 sns.stripplot(tips,x='day',y='total_bill')
```

```python
# cat plot -> figure level
sns.catplot(tips,x='day',y='total_bill',kind='strip')
```

```python
# jitter
sns.catplot(tips,x='day',y='total_bill',kind='strip',jitter=0.2,hue='sex')
```

```python
# swarm plot
sns.catplot(tips,x='day',y='total_bill',kind='swarm',hue='sex')
```

CATEGORICAL DISTRIBUTION PLOT

```python
# box plot
```

```python
sns.boxplot(tips,x='day',y='total_bill')
```

```python
# in fig
sns.catplot(tips,x='sex',y='total_bill',kind='box',hue='day')
```

```python
sns.boxplot(tips,y='total_bill')
```

```python
# VIOLIN PLOT -> (BOX PLOT + KDEplot)
```

```python
sns.violinplot(tips,x='day',y='total_bill')
```

```python
from os.path import split
sns.catplot(tips,x='day',y='total_bill',kind='violin',hue='sex',split=True)
```

CATEGORICAL ESTIMATE PLOT

```python
import numpy as np
```

```python
# BAR PLOT calcualte average bill on sex basis
sns.barplot(tips,x='sex',y='total_bill',hue='smoker',estimator=np.max)
```

```python
# point plot
```

```python
sns.barplot(tips,x='sex',y='total_bill',estimator=np.min,hue='smoker',errorbar=None)
```

```python
sns.pointplot(tips,x='sex',y='total_bill',estimator=np.min,hue='smoker',errorbar=None)
```

```python
# count plot
```

```python
sns.countplot(tips,x='sex',hue='day')
```

```python
sns.catplot(tips,x='sex',y='total_bill',kind='box',col='sex',hue='smoker',row='time')
```

Reg plot

```python
# axis level , it doesnt has hue
sns.regplot(tips,x='total_bill',y='tip')
```

```python
# figure level has hue
sns.lmplot(tips,x='total_bill',y='tip',hue='sex')
```

residplot

```python
sns.residplot(tips,x='total_bill',y='tip')
```

Asecond way to plot Facet plot -> facet Grid

```python
# figure plot ->relplot -> displot -> catplot -> lmplot

```

```python
sns.catplot(tips,x='sex',y='total_bill',kind='violin',col='day')
```

```python
g = sns.FacetGrid(tips,col='day',row='time',hue='smoker')
g.map(sns.violinplot,'sex','total_bill')
```

PLOTTING PAIR WISE RELATION (PAIR GRID VS PAIR PLOT)

PAIR PLOT

```python
iris
```

```python
sns.pairplot(iris,hue='species')
```

PAIR GRID

```python
g = sns.PairGrid(iris,hue='species')
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(iris,hue='species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(iris,hue='species')
g.map_diag(sns.violinplot)
g.map_offdiag(sns.lineplot)
```

```python
g = sns.PairGrid(iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

JOINT PLOT

```python
sns.jointplot(tips,x='total_bill',y='tip',kind='scatter')
```

```python
sns.jointplot(tips,x='total_bill',y='tip',kind='kde',hue='sex')
```

JOINT GRID

```python
g = sns.JointGrid(tips,x='total_bill',y='tip')
g.plot(sns.scatterplot,sns.histplot)
```

```python
g = sns.JointGrid(tips,x='total_bill',y='tip')
g.plot(sns.kdeplot,sns.violinplot)
```
