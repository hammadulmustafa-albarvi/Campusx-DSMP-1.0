### **Problem-1:** Make a subplots which have 2 plots.

- For the first chart, draw a scatter plot "*Monitored Cap.(MW)*" vs "*Total Cap. Under Maintenance (MW)*" of top 5 most frequent power stations. Then draw the lines which indicate the average values of these two columns. Change the colors according to the names of the Power Stations.
- For the second chart, draw a scatter plot "*Monitored Cap. (MW)*" vs "*Actual(MU)*" of the top 5 most frequent power stations. Also draw the lines which indicates the average values of these two columns. Change the colors according to the names of the Power Stations.


````python
df = pd.read_csv("https://rb.gy/gsmddu")
````

````python
plot1 = df[df['Power Station'].isin(df['Power Station'].value_counts().head().index)]
````

````python
plot1['color'] = plot1['Power Station'].replace({'NTPC Ltd.':0,'NHPC':1,'NPCIL':2,'GMR ENERG':3,'APL':4})
````

````python
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(15,7))
ax[1].scatter(plot1['Monitored Cap.(MW)'],plot1['Total Cap. Under Maintenace (MW)'],c=plot1['color'])
ax[1].axvline(plot1['Monitored Cap.(MW)'].mean(),color='red')
ax[1].axhline(plot1['Total Cap. Under Maintenace (MW)'].mean(),color='green')
ax[1].set_title('Scatter Plot')
ax[1].set_xlabel('Monitored Cap.(MW)')
ax[1].set_ylabel('Total Cap. Under Maintenace (MW)')

ax[0].set_title('Scatter Plot')
ax[0].set_xlabel('Monitored Cap.(MW)')
ax[0].set_ylabel('Actual(MU)')
ax[0].scatter(plot1['Monitored Cap.(MW)'],plot1['Actual(MU)'],c=plot1['color'])
ax[0].axvline(plot1['Monitored Cap.(MW)'].mean(),color='red')
ax[0].axhline(plot1['Actual(MU)'].mean(),color='green')
````

### **Problem-3:** Make a 3D *Surface* plot of this below mathematical equation.

$$z = |x| - |y|$$

````python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx,yy = np.meshgrid(x,y)
z = abs(xx)-abs(yy)
fig = plt.figure(figsize=(15,7))
ax = plt.subplot(projection='3d')
ax.plot_surface(xx,yy,z,cmap='viridis')
````

### **Problem-4:** Draw the 3D *Contour plot* of this below equation:

$$z = |x| - |y|$$

````python
fig = plt.figure(figsize=(12,6))
ax = plt.subplot()
ax.contourf(xx,yy,z,cmap='viridis')
````

### `Problem-6` Use Pandas plot functions

* Line plot of closing value of top 5 Stocks in Year 2020.
* Take top 5 stocks based on total turnover in Year 2020

````python
df = pd.read_csv('/content/nifty-50 - nifty-50.csv')
````

````python
df['Date'] = pd.to_datetime(df['Date'])
````

````python
df['Symbol'].value_counts()
````

````python
plot2 = df[df['Date'].dt.year == 2020]
````

````python
plot2.groupby('Symbol')['Turnover'].sum().sort_values(ascending=False).head().index
````

````python
top_5 =  plot2[plot2['Symbol'].isin(plot2.groupby('Symbol')['Turnover'].sum().sort_values(ascending=False).head().index)]
````

````python
top_5['color'] = top_5['Symbol'].replace({'RELIANCE':0,'AXISBANK':1,'HDFC':2,'BAJFINANCE':3,'HDFCBANK':4})
````

````python
# Index(['RELIANCE', 'AXISBANK', 'HDFC', 'BAJFINANCE', 'HDFCBANK'], dtype='object', name='Symbol')
plt.plot(top_5[top_5['color']==0]['Date'],top_5[top_5['color']==0]['Close'],label='RELIANCE')
plt.plot(top_5[top_5['color']==1]['Date'],top_5[top_5['color']==1]['Close'],label='AXISBANK')
plt.plot(top_5[top_5['color']==2]['Date'],top_5[top_5['color']==2]['Close'],label='HDFC')
plt.plot(top_5[top_5['color']==3]['Date'],top_5[top_5['color']==3]['Close'],label='BAJFINANCE')
plt.plot(top_5[top_5['color']==4]['Date'],top_5[top_5['color']==4]['Close'],label='HDFCBANK')
plt.legend()
````

### `Problem 9:` Create a surface plot and the 2 types of the contour plots of the below equation.

$$z = sin(\sqrt{x^2 + y^2})$$


````python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx,yy = np.meshgrid(x,y)
z = np.sin(np.sqrt(abs(xx)**2+abs(yy)**2))
fig = plt.figure(figsize=(15,7))
ax = plt.subplot(projection='3d')
ax.plot_surface(xx,yy,z,cmap='viridis')
````

````python
ax = plt.subplot()
ax.contour(xx,yy,z,cmap='viridis')
````

````python
ax = plt.subplot()
ax.contourf(xx,yy,z,cmap='viridis')
````

### `Problem 10:` Create a surface plot and the 2 types of the contour plots of the below equation.

$$z = tan(\log_2({x^2 + y^2})$$

````python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx,yy = np.meshgrid(x,y)
z = np.tan(np.log2(abs(xx)**2+abs(yy)**2))
fig = plt.figure(figsize=(15,7))
ax = plt.subplot(projection='3d')
ax.plot_surface(xx,yy,z,cmap='viridis')
````

````python
ax = plt.subplot()
ax.contour(xx,yy,z,cmap='viridis')
````

````python
ax = plt.subplot()
ax.contourf(xx,yy,z,cmap='viridis')
````
