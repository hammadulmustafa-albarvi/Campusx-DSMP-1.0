```python
# lines style
# line plot color
# labels titile
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dotted',linewidth=3,marker='D')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dotted',linewidth=2,marker='D')
plt.title("R Sharma vs V kohli career comparision")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.show()
```

```python
#marker and its size
# lines style
# line plot color
# labels titile
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dotted',linewidth=3,marker='D',markersize=19)
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dotted',linewidth=2,marker='D')
plt.title("R Sharma vs V kohli career comparision")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.show()
```

```python
#lables for vkohli and sharma
#marker and its size
# lines style
# line plot color
# labels titile
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dotted',linewidth=3,marker='D',markersize=10,label='V Kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dotted',linewidth=2,marker='D',label='R Sharma')
plt.title("R Sharma vs V kohli career comparision")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend()
plt.show()
```

```python
# vertical bar

children=[10,20,40,5,23]
colors = ['red','blue','green','yellow','pink']
plt.bar(colors,children,color='red')
```

```python
 # horizontal bar

children=[10,20,40,5,23]
colors = ['red','blue','green','yellow','pink']
plt.barh(colors,children,color='red')
```

```python
df =  pd.read_csv("/content/batsman_season_record.csv")
df
```

```python
plt.bar(df['batsman'],df['2015'])
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2 )
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2) 
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
# rotate x ticks

# vertical bar

children=[10,20,40,5,23]
colors = ['redredredredredredvred','blueredvvredred','greenredredred','yellowredredvred','pinkredvvvred']
plt.bar(colors,children,color='red')
plt.xticks(rotation='vertical')
```

```python
# stacked bar chart

plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016') 
plt.bar(df['batsman'],df['2017'],bottom=df['2016']+df['2015'],label='2017')
plt.legend()
```
