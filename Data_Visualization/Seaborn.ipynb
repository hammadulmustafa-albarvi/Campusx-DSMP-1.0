### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips= sns.load_dataset('tips')
```

```python
sns.scatterplot(tips,x='total_bill',y='tip',hue='sex',style='time',size='size')
```

```python
sns.relplot(tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')
```

```python
gap = px.data.gapminder()
```

```python
tmp_df = gap[gap['country']=='Pakistan']
```

```python
sns.lineplot(tmp_df,x='year',y='lifeExp')
```

```python
sns.relplot(tmp_df,x='year',y='lifeExp',kind='line')
```

```python
# hue
tmp_df = gap[gap['country'].isin(['Germany','Brazil','China'])]
```

```python
sns.relplot(tmp_df,x='year',y='lifeExp',kind='line',hue='country',style='continent',size='continent')
```

```python
# facet plot it will work with rel plot only and not with axes level scatter/line plot that we call by sns.scatter/sns.line
sns.relplot(tips,x='total_bill',y='tip',kind='scatter',col='sex',row='day')
```

```python
# COL wrap
sns.relplot(gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year',col_wrap=3)
```

### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

```python
 # figure level-> displot
 # axis level -> histplot,kdeplot,rugplot
```

```python
sns.histplot(tips,x='total_bill')
```

```python
sns.displot(tips,x='total_bill',kind='hist')
```

```python
sns.displot(tips,x='total_bill',kind='hist',bins=5)
```

```python
# plot histogram on catergorical

sns.displot(tips,x='day',kind='hist')

# its acting like a count plot but not pure histogram
```

```python
sns.displot(tips,x='tip',kind='hist',hue='sex')
```

```python
sns.displot(tips,x='tip',kind='hist',hue='sex',element='step')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
sns.displot(titanic,x='age',hue='sex')
```

```python
# col wont work on axes level function it works only on fig level

sns.displot(titanic,x='age',col='sex')
```

KDE PLOT

```python
# kde plot
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate
```

```python
sns.kdeplot(tips,x='total_bill',hue='sex',fill=True)
```

```python
sns.displot(tips,x='total_bill',kind='kde',height=7,aspect=2)
```

```python
 #Rugplot

# Plot marginal distributions by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
```

```python
sns.kdeplot(tips,x='total_bill')
sns.rugplot(tips,x='total_bill')
```

```python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot
# and then shows the count of observations within each rectangle with the fill color


sns.histplot(tips,x='total_bill',y='tip')
```

```python
sns.displot(tips,x='total_bill',y='tip',kind='hist')
```

```python
# Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian
```

```python
sns.displot(tips,x='total_bill',y='tip',kind='kde')
```

### 2. Matrix Plot

- Heatmap
- Clustermap

```python
# heatmap
```

```python
tmp_df = gap.pivot(index='country',columns='year',values='lifeExp')
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(tmp_df)
```

```python
tmp_df = gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')
```

```python
tmp_df.head()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(tmp_df,annot=True,linewidths=1,cmap='viridis')
```

Cluster Map

```python
# Clustermap

# Plot a matrix dataset as a hierarchically-clustered heatmap.

# This function requires scipy to be available.


iris = px.data.iris()
```

```python
iris.head(1)
```
