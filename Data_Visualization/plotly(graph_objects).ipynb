```python
import numpy as np
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
pd.set_option('display.max_columns', None)
```

```python
match = pd.read_csv('/content/matches.csv')
delivery = pd.read_csv('/content/deliveries.csv')
```

```python
ipl = delivery.merge(match,left_on='match_id',right_on='id')
```

Scatter Plot between batsman average and strike rate of top 50 batsman

```python
top_50 = ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(50).index.tolist()
```

```python
tmp_df = ipl[ipl['batsman'].isin(top_50)]
```

```python
runs = tmp_df.groupby('batsman')['batsman_runs'].sum()
balls = tmp_df.groupby('batsman')['batsman_runs'].count()
strike_rate = (runs/balls) * 100
strike_rate = strike_rate.reset_index()
```

```python

out = tmp_df['player_dismissed'].value_counts()
average = runs/out
average = average.reset_index()
average = average.rename(columns={'index':'batsman',0:'avg'})
```

```python
average = average.merge(strike_rate,on='batsman')
```

```python
trace = go.Scatter(x=average['avg'],y=average['batsman_runs'],mode='markers',text=average['batsman'],
                   marker={'color':'red','size':16})
data = [trace]
layout = go.Layout(title='Scatter plot')
fig = go.Figure(data=trace,layout=layout)
fig.show()
```

year by year line chart of performance

```python
yby_performance=ipl[ipl['batsman']=='V Kohli'].groupby('season')['batsman_runs'].sum().reset_index()
yby_performance1=ipl[ipl['batsman']=='MS Dhoni'].groupby('season')['batsman_runs'].sum().reset_index()
```

```python
def batsman_comp(*names):
  data = []
  for i in names:
    performance=ipl[ipl['batsman']==i].groupby('season')['batsman_runs'].sum().reset_index()
    trace = go.Scatter(x=performance['season'],y=performance['batsman_runs'],name=i)
    data.append(trace)
  fig = go.Figure(data)
  fig.show()
```

```python
batsman_comp('V Kohli','MS Dhoni','RG Sharma','DA Warner')
```

Histogram

```python
balls = ipl.groupby('batsman')['batsman_runs'].count()
```

```python
tmp_df = ipl.merge(balls,on='batsman')
```

```python
tmp_df = tmp_df[tmp_df['batsman_runs_y'] > 150]
```

```python
runs = tmp_df.groupby('batsman')['batsman_runs_x'].sum()
balls = tmp_df.groupby('batsman')['batsman_runs_x'].count()
strike_rate = (runs/balls) * 100
strike_rate = strike_rate.reset_index()
```

```python
strike_rate
```
