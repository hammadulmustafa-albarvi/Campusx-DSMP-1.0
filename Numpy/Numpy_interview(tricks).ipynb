Plotting 3D graphs


meshgrid

```python
import numpy as np
import pandas as pd
```

```python
a = np.linspace(-10,9,20)
b = np.linspace(-10,9,20)
```

```python
x,y = np.meshgrid(a,b)
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
def func(x,y):
  return x**2 + y**2
z = func(x,y)
z
```

```python
import plotly.express as px
import plotly.graph_objects as go
fig = px.scatter_3d()
fig.add_trace(go.Surface(x=x,y=y,z=z))
fig.show()
```
