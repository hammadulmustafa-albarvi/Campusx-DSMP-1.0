```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('/content/pca_dataset.csv')
```

```python
df.head()
```

```python
sc = StandardScaler()
df.iloc[:,0:3] = sc.fit_transform(df.iloc[:,0:3])
```

```python
covariance_matrix = np.cov([df.iloc[:, 0], df.iloc[:, 1],df.iloc[:,2]])
```

```python
covariance_matrix
```

```python
eigen_values,eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
eigen_values
```

```python
eigen_vectors
```

```python
np.dot(eigen_values[0] , eigen_vectors[0][0])
```

```python
res = np.dot(df.iloc[:,0:3],eigen_vectors[:,0].T)
```

```python
res = pd.DataFrame(res,columns=['PC'])
res['target'] = df['Target']
```
