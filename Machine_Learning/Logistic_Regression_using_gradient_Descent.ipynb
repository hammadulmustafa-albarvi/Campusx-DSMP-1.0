```python
import numpy as np
import pandas as pd
```

```python
class gradient_logistic_regression:
  def __init__(self,X,y,lr,epochs):
    self.lr = lr
    self.epochs = epochs
    self.X = np.insert(X,0,1,axis=1)
    self.weights = np.ones(self.X.shape[1])
    self.y = y

  def predict(self):
    for i in range(self.epochs):
      z = np.dot(self.X,self.weights)
      y_hat = 1/(1 + np.exp(-z))
      self.weights = self.weights + self.lr * (np.dot((self.y-y_hat),self.X)/self.X.shape[0])
    return self.weights[1:],self.weights[0]
```
