Data set Link - https://www.kaggle.com/datasets/altavish/boston-housing-dataset

DataSet description:
```
ZN: proportion of residential land zoned for lots over 25,000 sq.ft.

RM: average number of rooms per dwelling

```

````python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso,LinearRegression,Ridge,ElasticNet
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
````

````python
boston  = pd.read_csv('/content/HousingData.csv')
````

````python
boston
````

````python
boston.dropna(inplace=True)
````

````python
X = boston.iloc[:,:-1]
y = boston.iloc[:,-1]
````

````python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=20)
````

````python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
````

# Using simple Linear Regression

````python
lr = LinearRegression()
lr.fit(X_train,y_train)
r2_score(y_test,lr.predict(X_test))
````

# Using polynomial Regression

````python
polynomial = PolynomialFeatures(degree=2)
X_poly = polynomial.fit_transform(X_train)
X_test_poly = polynomial.transform(X_test)
poly_lr = LinearRegression()
poly_lr.fit(X_poly,y_train)
r2_score(y_test,poly_lr.predict(X_test_poly))
````

# Using Ridge Regression

## On normal data

````python
ridge = Ridge(alpha=0.0005)
ridge.fit(X_train,y_train)
r2_score(y_test,ridge.predict(X_test))
````

## On polynomial Data

````python
ridge = Ridge(alpha=1000000)
ridge.fit(X_poly,y_train)
r2_score(y_test,ridge.predict(X_test_poly))
````

# Using Lasso

## On linear/normal Data

````python
lasso = Lasso(alpha=0.00001)
lasso.fit(X_train,y_train)
r2_score(y_test,lasso.predict(X_test))
````

## On polynomial Data

````python
lasso = Lasso(alpha=0.001)
lasso.fit(X_poly,y_train)
r2_score(y_test,lasso.predict(X_test_poly))
````

# We can conclude the polynomial regression works better if we remove overfitting from it
