### `Solution`

```python
import pandas as pd

df = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vQtBXo5cBnDsM2fmfHPm6u72KGUS5FjPHNGMxOfYjA9-CAhmnRpwkIw_rOR3sANJIToiUU__6fbBvig/pub?gid=572763137&single=true&output=csv")

df.info()
```

```python
df.shape
```

Remove Duplicates

```python
df = df.loc[:,~df.T.duplicated()]
```

```python
df.fillna(df.mean(numeric_only=True),inplace=True)
```

Variance Threshold

```python
from sklearn.model_selection import train_test_split
```

```python
df['Time'] = pd.to_datetime(df['Time'])
```

```python
X  =df.iloc[:,1:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
log_reg = LogisticRegression(max_iter=1000)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=20)
```

```python
log_reg.fit(X_train,y_train)
y_pred = log_reg.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.feature_selection import VarianceThreshold
vt = VarianceThreshold(threshold=0.05)
```

```python
vt.set_output(transform='pandas')
```

```python
X_train.shape
```

```python
X_train = vt.fit_transform(X_train)
X_test = vt.transform(X_test)

```

```python
X_train.shape
```

Anova

```python
from sklearn.feature_selection import f_classif
from sklearn.feature_selection import SelectKBest

anova = SelectKBest(f_classif,k=100)
```

```python
anova.set_output(transform='pandas')
```

```python
X_new = anova.fit_transform(X_train,y_train)
X_test =  anova.transform(X_test)
```

```python
X_new.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
log_reg = LogisticRegression(max_iter=1000)
```

```python
y_train.shape
```

```python
log_reg.fit(X_new,y_train)
y_pred = log_reg.predict(X_test)
accuracy_score(y_test,y_pred)
```

Correlation

```python
correlation = X_train.corr()
```

```python
correlated_cols = []
for i in range(correlation.shape[1]):
  for j in range(i+1,correlation.shape[1]):
    if correlation.iloc[i,j] > 0.95: 
      correlated_cols.append(correlation.columns[i])
```

```python
X_train.drop(columns =correlated_cols )
```
