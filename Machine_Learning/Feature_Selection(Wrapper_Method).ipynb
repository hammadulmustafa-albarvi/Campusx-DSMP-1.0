```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression,LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
import pandas as pd
import seaborn as sns
import numpy as np
```

# Exhaustive Feature Selection

```python
df = sns.load_dataset('iris')
```

Logistic Regression

```python
df.head()
```

```python
from mlxtend.feature_selection import ExhaustiveFeatureSelector as EFS
```

```python
lr = LogisticRegression()
```

```python
sel = EFS(lr,max_features=4,scoring='accuracy',cv=5)
model = sel.fit(df.iloc[:,:4],df['species'])
```

```python
model.best_score_
```

```python
model.best_feature_names_
```

```python
model.subsets_
```

Linear Regression

```python
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv')
```

```python
df.head()
```

```python
X = df.iloc[:,0:-1]
y = df.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test =  train_test_split(X,y,test_size=0.2,random_state=20)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
lr = LinearRegression()
```

```python
np.mean(cross_val_score(lr,X_train,y_train,cv=5,scoring='r2'))
```

```python
np.mean(cross_val_score(lr,X_test,y_test,cv=5,scoring='r2'))
```

```python
lr = LinearRegression()
```

```python
efs = EFS(lr,max_features=13,scoring='r2',cv=10,print_progress=True,n_jobs=-1)
sel = efs.fit(X_train,y_train)
```

```python
sel.best_score_
```

```python
sel.best_feature_names_
```

```python
pd.DataFrame.from_dict(sel.get_metric_dict()).T
```

```python
X_train_sel = sel.transform(X_train)
X_test_sel = sel.transform(X_test)
```

```python
np.mean(cross_val_score(lr,X_train_sel,y_train,cv=5,scoring='r2'))
```

```python
np.mean(cross_val_score(lr,X_test_sel,y_test,cv=5,scoring='r2'))
```

# Sequential Backward Selection

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
sfs = SFS(lr,k_features='best',forward=False,scoring='r2',cv=5)
sfs.fit(X_train,y_train)
```

```python
sfs.k_score_
```

```python
X_train_sel = sfs.transform(X_train)
```

```python
X_test_sel = sfs.transform(X_test)
```

```python
np.mean(cross_val_score(lr,X_test_sel,y_test,cv=5,scoring='r2'))
```

```python
np.mean(cross_val_score(lr,X_train_sel,y_train,cv=5,scoring='r2'))
```
