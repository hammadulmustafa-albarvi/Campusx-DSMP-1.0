```python
import pandas as pd
```

```python
pd.crosstab(df['Outlook'],df['Play Tennis'],normalize='columns').stack().to_dict()
```

```python
ans = {}
for i in ['Outlook', 'Temperature', 'Humidity', 'Wind', 'Play Tennis']:
  ans.update(pd.crosstab(df[i],df['Play Tennis'],normalize='columns').stack().to_dict())
```
