```python
def convert_to_time(x):
  if len(x.split()) == 5:
    hours = x.split()[0]
    mins = x.split()[3]
    return pd.Timedelta(hours=int(hours),minutes=int(mins))
  elif len(x.split()) == 2:
    if x.split()[1] == 'hrs':
      return pd.Timedelta(hours=int(x.split()[0]))
    elif x.split()[1] == 'mins':
      return pd.Timedelta(minutes=int(x.split()[0]))
```

```python
df['total_time']=df['time'].apply(convert_to_time)
```

```python
df.head(2)
```

```python
def extract_reviews(x):
  if len(x.split()) > 3:
    review = x.split()[0]
    return review
  else:
    review = 0
    return review
```

```python
def extract_rating(x):
  if len(x.split()) > 3:
    rating = x.split()[4]
    return rating[5:]
  else:
    rating = 0
    return rating
```

```python
df['reviews'] = df['stars'].apply(extract_reviews)
```

```python
df['rating'] = df['stars'].apply(extract_rating)
```
