```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
pd.set_option('display.max_rows',None)
```

```python
# soup = BeautifulSoup(webpage,'lxml')
```

```python
final_df = pd.DataFrame()
for i in range(1,25):
    url = f'https://www.scrapethissite.com/pages/forms/?page_num={i}'
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get(url,headers=headers).text

    soup = BeautifulSoup(webpage,'lxml')

    team_name = [] 
    year = []
    wins = []
    losses = []
    win_percentage = []
    goals_for = []
    goals_against = []
    goal_differential = [] 

    team = soup.find_all('tr',class_='team')
    
    for i in team:
        team_name.append(i.find('td',class_='name').text.strip())
        year.append(i.find('td',class_='year').text.strip())
        wins.append(i.find('td',class_='wins').text.strip())
        losses.append(i.find('td',class_='losses').text.strip())
        
        # win_percentage.append(i.find('td',class_='pct text-success').text.strip())
    
    
        if i.find('td',class_='pct text-success')==None:
            win_percentage.append(i.find('td',class_='pct text-success'))
        else:
            win_percentage.append(i.find('td',class_='pct text-success').text.strip())
        
        
        goals_for.append(i.find('td',class_='gf').text.strip())
        goals_against.append(i.find('td',class_='ga').text.strip())
    
        
        # goal_differential.append(i.find('td',class_='diff text-success').text.strip())
    
        if i.find('td',class_='diff text-success')==None:
            goal_differential.append(i.find('td',class_='diff text-success'))
        else:
            goal_differential.append(i.find('td',class_='diff text-success').text.strip()) 
    
    d = {'Team':team_name,'Year':year,'Wins':wins,'Losses':losses,'Win%':win_percentage,'Goals For':goals_for,'Goals Against':goals_against,'+/-':goal_differential}
    df = pd.DataFrame(d)
    final_df = pd.concat([final_df,df],ignore_index=True)
```
