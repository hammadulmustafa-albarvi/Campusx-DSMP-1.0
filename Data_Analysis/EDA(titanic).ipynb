```python
df['individual_fare']=df['Fare']/(df['SibSp']+df['Parch']+1)
```

```python
df['individual_fare'].plot(kind='box')
```

```python
df[['Fare','individual_fare']].describe()
```

```python
df['family_size'] = df['SibSp']+df['Parch']+1
```

```python
df.head()
```

```python
def transform_family_size(x):
  if x==1:
    return 'alone'
  elif x>1 and x<5:
    return 'small'
  else:
    return 'large'
```

```python
df['family_type'] = df['family_size'].apply(transform_family_size)
```

```python
df.head()
```

```python
pd.crosstab(df['Survived'],df['family_type'],normalize='columns')
```

```python
df['sur_name'] = df['Name'].str.split(',').str.get(0)
```

```python
df.head()
```

```python
df['title'] = df['Name'].str.split(',').str.get(1).str.strip().str.split(' ').str.get(0)
```

```python
df.head()
```

```python
df['title'].value_counts()
```

```python
df['title']=df['title'].str.replace('Rev.','other').replace('Dr.','other').replace('Col.','other').replace('Major.','other').replace('Capt.','other').replace('the','other').replace('Jonkheer.','other')
```

```python
df['title'].value_counts()
```

```python
pd.crosstab(df['Survived'],df['title'],normalize='columns')
```
