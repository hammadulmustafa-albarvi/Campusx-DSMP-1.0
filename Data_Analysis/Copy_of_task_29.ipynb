# Feature Engineering

```python
def children_amount(x):
  if x==0:
    return 0
  else:
    return 1
```

```python
df['has_children'] = df['children'].apply(children_amount)
```

```python
df.sample(5)
```

 ## Has children and claim

 number of claims for people who have children are more than people with no children

 mean claim of people with no children is less than people with children

 but the highest claim is made by person who has no children

```python
df.groupby('has_children')['claim'].count().plot(kind='bar')
```

```python
sns.barplot(df,x='has_children',y='claim',estimator=np.max)
```

```python
sns.violinplot(df,x='has_children',y='claim')
```

```python
def age_bins(x):
  if x >= 18 and x < 30:
    return 'young'
  elif x >= 30 and x < 50:
    return 'middle_aged'
  elif x>=50 and x<=60:
    return 'senior_citizen'
  else:
    return 'missing age'
```

```python
df['age_type'] = df['age'].apply(age_bins)
```

## age type and claim

conclusions

more middle aged people have made more claims

mean claims made by young people are the largest

highest claims are made by middle aged

```python
df['age_type'].value_counts().plot(kind='bar')
```

```python
sns.barplot(df,x='age_type',y='claim',estimator=np.mean)
```

```python
sns.boxplot(df,x='age_type',y='claim')
```

```python
df['bmi'].sort_values()
```

```python
def weight_category(x):
  if x < 18.5:
    return 'under_weight'
  elif x>=18.5 and x<25:
    return 'normal_weight'
  elif x>=25 and x<30:
    return 'overweight_1'
  elif x>=30 and x<35:
    return 'overweight_2'
  elif x>=35 and x<40:
    return 'overweight_3'
  else:
    return 'overweight_4'
```

```python
df['weight_type'] = df['bmi'].apply(weight_category)
```

```python
df.head()
```

```python
df['bloodpressure'].sort_values()
```

```python
def blood_pressure_range(x):
  if x < 120:
    return 'normal'
  elif x >= 120 and x< 130:
    return 'pre_hypertension'
  elif x>=130 and x<140:
    return 'hypertension_stage_1'
  else:
    return 'hypertension_stage_2'
```

```python
df['bloodpressure_type'] = df['bloodpressure'].apply(blood_pressure_range)
```
