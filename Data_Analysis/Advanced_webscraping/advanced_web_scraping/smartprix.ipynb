```python
with open('smartprix.html','r',encoding='utf-8') as f:
    html=f.read()
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html,'lxml')
```

```python
len(soup.find_all('div',class_='sm-product has-tag has-features has-actions'))
```

```python
import pandas as pd
import numpy as np
name=[]
rating=[]
price=[]
sim=[]
processor=[]
ram=[]
battery=[]
display=[]
camera=[]
card=[]
os=[]


for i in soup.find_all('div',class_='sm-product has-tag has-features has-actions'):
    data=[]
    name.append((i.find('h2').text))
    price.append((i.find('span',class_='price').text))
    rating.append((i.find('b').text))
    for j in (i.find('ul',class_='sm-feat specs')):
        data.append(str(j.text))
    try:
        sim.append(data[0])
    except:
        sim.append(np.nan)


    try:
        processor.append(data[1])
    except:
        processor.append(np.nan)


    try:
        ram.append(data[2])
    except:
        ram.append(np.nan)

    try:
        battery.append(data[3])
    except:
        battery.append(np.nan)

    try:
        display.append(data[4])
    except:
        display.append(np.nan)

    try:
        camera.append(data[5])
    except:
        camera.append(np.nan)

    try:
        card.append(data[6])
    except:
        card.append(np.nan)

    try:
        os.append(data[7])
    except:
        os.append(np.nan)
        
    # ram.append(data[2])
    # battery.append(data[3])
    # display.append(data[4])
    # camera.append(data[5])
    # card.append(data[6])
    # os.append(data[7])
    
d = {'name':name,'rating':rating,'price':price,'sim':sim,'processor':processor,'ram':ram,'battery':battery,'display':display,'camera':camera,'card':card,'os':os}
df = pd.DataFrame(d)
```
